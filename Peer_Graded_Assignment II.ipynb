**Question 2: Use Webscraping to Extract Tesla Revenue Data**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website containing Tesla revenue data
url = 'https://example.com/tesla-revenue-page'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Identify the HTML elements containing the revenue data
# Extract and store the data in a list or dictionary

# Example:
revenue_data = {'Date': ['2022-01-01', '2022-02-01', '2022-03-01'],
                'Revenue': [1000000, 1500000, 1200000]}

# Create a DataFrame from the extracted data
tesla_revenue = pd.DataFrame(revenue_data)

# Display the last five rows
print(tesla_revenue.tail())
```

**Question 5: Plot Tesla Stock Graph**

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    # Assuming data has 'Date' and 'Close' columns
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Price', marker='o')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Assuming tesla_data has been previously loaded
make_graph(tesla_data, 'Tesla Stock Price Over Time')
```
